# Upload files from a folder (Single files)






```python
#@title Singular File Uploads
#@markdown Please ensure that you have noted where you are uploading your files in the "REPO" section. Follow these steps to create a new repository on your Hugging Face account:
#@markdown [Navigate to Your Hugging Face Account](https://huggingface.co/)

#@markdown Visit the Hugging Face website and log in to your account via the above link

#@markdown Go to your user profile page.
#@markdown [Click on the "New  Model" button.](https://huggingface.co/new)

#@markdown Enter a name for your model repository.
#@markdown Choose the appropriate settings and permissions.

#@markdown Note the name of the newly created repository.

#@markdown You'll need to ask ChatGPT for the updated widget, which you just hit the button and everything is mostly straight forward. You just add your drive/storage information, User information and follow the boxes.

import glob
import os
from huggingface_hub import HfApi
from ipywidgets import *
from IPython.display import display
from pathlib import Path

# Define widgets for user inputs
drive_path = Text(value='/content/drive/My Drive', placeholder='Path in Google Drive', description='Drive Path:')
hfuser = Text(placeholder='YourUSERHERE', description='HF User:')
hfrepo = Text(placeholder='ModelNameHere', description='HF Repo:')
update_btn = Button(description='Update Files')
upload_btn = Button(description='Upload')
out = Output()

# Initialize Hugging Face API
api = HfApi()

# Define function to update file list
def update_files(_):
    global ckpt_picker
    path = Path(drive_path.value)
    if path.exists() and path.is_dir():
        all_ckpts = [str(f) for f in path.glob("*.safetensors")]
        ckpt_picker.options = all_ckpts
    else:
        with out:
            print("Invalid directory. Please check the path and try again.")

# Define upload function
def upload_ckpts(_):
    repo_id = f"{hfuser.value or hfuser.placeholder}/{hfrepo.value or hfrepo.placeholder}"
    with out:
        if ckpt_picker is None or len(ckpt_picker.value) < 1:
            print("Nothing selected for upload, make sure to select one of the checkpoint files in the list, or verify there are checkpoint files in the specified directory.")
        for ckpt in ckpt_picker.value:
            print(f"Uploading to HF: huggingface.co/{repo_id}/{os.path.basename(ckpt)}")
            response = api.upload_file(
                path_or_fileobj=ckpt,
                path_in_repo=os.path.basename(ckpt),
                repo_id=repo_id,
                repo_type=None,
                create_pr=1,
                commit_message="Upload with Earth & Dusk Huggingface ðŸ¤— Backup"
            )
            display(response)
        print("DONE")
        print("Go to your repo and accept the PRs created to see your files.")

# Bind the functions to buttons
update_btn.on_click(update_files)
upload_btn.on_click(upload_ckpts)

# Create and display the widget layout
ckpt_picker = SelectMultiple(options=[], layout=Layout(width="600px"))
box = VBox([
    drive_path,
    HBox([hfuser, hfrepo]),
    update_btn,
    ckpt_picker,
    upload_btn,
    out
])

display(box)
```
