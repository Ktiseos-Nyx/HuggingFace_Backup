## Upload files from a folder (Single files)
-----------------


Please ensure that you have noted where you are uploading your files in the "REPO" section. Follow these steps to create a new repository on your Hugging Face account:

[Navigate to Your Hugging Face Account](https://huggingface.co/)

Visit the Hugging Face website and log in to your account via the above link

Go to your user profile page.
[Click on the "New  Model" button.](https://huggingface.co/new)
Enter a name for your model repository.
Choose the appropriate settings and permissions.


Note the name of the newly created repository for use in your Jupyter Lab setup. Unlike Colab, which can automatically create a repository, Jupyter Lab requires a pre-created repository. This setup should allow you to upload all necessary files to your repository seamlessly.




```python
import glob
import os
from huggingface_hub import HfApi
from ipywidgets import *
from IPython.display import display
from pathlib import Path

# Define widgets for user inputs
file_type = Text(value='safetensors', placeholder='File Type (e.g., safetensors)', description='File Type:')
sort_by = Dropdown(options=['name', 'date'], value='name', description='Sort By:')
hfuser = Text(placeholder='EarthNDusk', description='HF User:')
hfrepo = Text(placeholder='July_2024_Backup_Full_Models', description='HF Repo:')
update_btn = Button(description='Update Files')
upload_btn = Button(description='Upload')
out = Output()

# Initialize Hugging Face API
api = HfApi()

# Define function to update file list
def update_files(_):
    global ckpt_picker
    file_extension = file_type.value
    all_ckpts = [f for f in glob.glob(f"*.{file_extension}")]
    
    if sort_by.value == 'date':
        all_ckpts.sort(key=os.path.getmtime)
    else:
        all_ckpts.sort()

    ckpt_picker.options = all_ckpts

# Define upload function
def upload_ckpts(_):
    repo_id = f"{hfuser.value or hfuser.placeholder}/{hfrepo.value or hfrepo.placeholder}"
    with out:
        if ckpt_picker is None or len(ckpt_picker.value) < 1:
            print("Nothing selected for upload, make sure to select one of the checkpoint files in the list, or verify there are checkpoint files in the specified directory.")
        for ckpt in ckpt_picker.value:
            print(f"Uploading to HF: huggingface.co/{repo_id}/{os.path.basename(ckpt)}")
            response = api.upload_file(
                path_or_fileobj=ckpt,
                path_in_repo=os.path.basename(ckpt),
                repo_id=repo_id,
                repo_type=None,
                create_pr=1,
                commit_message="Upload with Earth & Dusk Huggingface ðŸ¤— Backup"
            )
            display(response)
        print("DONE")
        print("Go to your repo and accept the PRs created to see your files.")

# Bind the functions to buttons
update_btn.on_click(update_files)
upload_btn.on_click(upload_ckpts)

# Create and display the widget layout
ckpt_picker = SelectMultiple(options=[], layout=Layout(width="600px"))
box = VBox([
    file_type,
    sort_by,
    HBox([hfuser, hfrepo]),
    update_btn,
    ckpt_picker,
    upload_btn,
    out
])

display(box)
```
